## Preprocessing

```python
import isodate
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

```python
useless_columns = ["Description", "RecipeYield", "RecipeServings", 'ReviewCount', 'AggregatedRating', 'CookTime']
recipes.drop(columns=useless_columns, inplace=True)
```

```python
np.array(recipes.columns)
```

```python
features = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent', 'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']
```

```python
max_Calories = 2000
max_daily_fat = 78
max_daily_Saturatedfat = 20
max_daily_Cholesterol = 300
max_daily_Sodium = 2300
max_daily_Carbohydrate = 275
max_daily_Fiber = 28
max_daily_Sugar = 50
max_daily_Protein = 50

max_list = [max_Calories, max_daily_fat, max_daily_Saturatedfat, max_daily_Cholesterol, max_daily_Sodium, max_daily_Carbohydrate, max_daily_Fiber, max_daily_Sugar, max_daily_Protein]
```

```python
filtered_data = recipes.copy()
columns = filtered_data[features].columns
for col, maximum in zip(columns, max_list):
    filtered_data = filtered_data[filtered_data[col] < maximum]
```

```python
filtered_data.shape
```

```python
def remove_outliers(column: str) -> pd.DataFrame:
    Q1 = filtered_data[column].quantile(0.25)
    Q3 = filtered_data[column].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return filtered_data[(filtered_data[column] >= lower_bound) & (filtered_data[column] <= upper_bound)]
```

```python
for col in filtered_data[features].columns:
    filtered_data = remove_outliers(col)
```

```python
filtered_data.shape
```

```python
meal_column = 'RecipeCategory'

# Count meal types before and after filtering
original_counts = recipes[meal_column].value_counts(normalize=True) * 100
filtered_counts = filtered_data[meal_column].value_counts(normalize=True) * 100

# Combine into one DataFrame for comparison
comparison = pd.DataFrame({
    'Before Filtering (%)': original_counts,
    'After Filtering (%)': filtered_counts
}).fillna(0).sort_values('Before Filtering (%)', ascending=False)

# Optional: calculate the drop in percentage
comparison['Change (%)'] = comparison['After Filtering (%)'] - comparison['Before Filtering (%)']

# Display
print(comparison.head(15))
```

```python
filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].fillna('Unknown')
```

```python
filtered_data['RecipeCategory'].isnull().sum()
```

```python
categories = list(filtered_data['RecipeCategory'].unique())
```

```python
breakfast_keywords = ['breakfast', 'scones', 'breads', 'bread', 'smoothies', 'cheese', 'pancake', 'egg', 'avocado', 'biscuit', 'cereal', 'muffin', 'fruit', 'jam', 'butter', 'oat', 'yogurt']
lunch_keywords = ['sandwich', 'salad', 'soup', 'wrap', 'rice', 'noodles', 'chowders', 'manicotti', 'beans', 'lentil', 'quiche', 'tofu', 'tuna']
dinner_keywords = ['chicken', 'pork', 'beef', 'stew', 'lamb', 'veal', 'fish', 'roast', 'spaghetti', 'meatballs', 'meatloaf', 'gravy', 'curry', 'casserole']

breakfast = []
lunch = []
dinner = []
finer_other = {
    'desserts': [],
    'ingredients': [],
    'cuisine': [],
    'appliance': [],
    'diet': [],
    'holiday/event': [],
    'time': [],
    'misc': []
}

# Helper function to classify categories
def classify_category(cat):
    cat_lower = cat.lower()
    if any(kw in cat_lower for kw in breakfast_keywords):
        return 'breakfast'
    elif any(kw in cat_lower for kw in lunch_keywords):
        return 'lunch'
    elif any(kw in cat_lower for kw in dinner_keywords):
        return 'dinner'
    elif any(kw in cat_lower for kw in ['cookie', 'cake', 'dessert', 'pie', 'ice cream', 'pudding', 'candy']):
        return 'desserts'
    elif any(kw in cat_lower for kw in ['chicken', 'beef', 'pork', 'rice', 'vegetable', 'fruit', 'tofu', 'bread', 'egg', 'fish', 'potato']):
        return 'ingredients'
    elif any(kw in cat_lower for kw in ['asian', 'european', 'mexican', 'thai', 'greek', 'french', 'italian', 'german', 'indian', 'cuisine']):
        return 'cuisine'
    elif any(kw in cat_lower for kw in ['oven', 'microwave', 'stove', 'pressure cooker', 'appliance', 'machine']):
        return 'appliance'
    elif any(kw in cat_lower for kw in ['low', 'high', 'free', 'vegan', 'kosher', 'lactose', 'cholesterol']):
        return 'diet'
    elif any(kw in cat_lower for kw in ['christmas', 'thanksgiving', 'birthday', 'memorial', 'holiday', 'event', 'halloween']):
        return 'holiday/event'
    elif any(kw in cat_lower for kw in ['<', 'mins', 'hours', 'quick']):
        return 'time'
    else:
        return 'misc'

# Classify each category
for cat in categories:
    group = classify_category(cat)
    if group == 'breakfast':
        breakfast.append(cat)
    elif group == 'lunch':
        lunch.append(cat)
    elif group == 'dinner':
        dinner.append(cat)
    else:
        finer_other[group].append(cat)
```

```python
desserts = finer_other['desserts']
ingredients = finer_other['ingredients']
cuisine = finer_other['cuisine']
diet = finer_other['diet']
holidays_events = finer_other['holiday/event']
time = finer_other['time']
misc = finer_other['misc']
appliance = finer_other['appliance']
misc.extend(appliance)
```

```python
for cat in categories:
    lower_cat = cat.lower()
    if cat in desserts:
        filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Dessert')
    elif cat in ingredients:
        filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Ingredients')
    elif cat in cuisine:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Cuisine')
    elif cat in diet:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Diet')
    elif cat in holidays_events:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Holiday/Event')
    elif cat in time:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Time')
    elif cat in misc:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Misc')
    elif cat in breakfast:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Breakfast')
    elif cat in lunch:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Lunch')
    elif cat in dinner:
            filtered_data['RecipeCategory'] = filtered_data['RecipeCategory'].replace(cat, 'Dinner')
```

```python
import re
def parse_r_vector(r_str):
    cleaned = re.sub(r'^c\(|\)$', '', r_str.strip())
    parts = re.findall(r'"(.*?)"', cleaned)
    return parts
```

```python
filtered_data['RecipeIngredientParts'] = filtered_data['RecipeIngredientParts'].apply(parse_r_vector)
filtered_data['RecipeIngredientQuantities'] = filtered_data['RecipeIngredientQuantities'].apply(parse_r_vector)
```

## Training the model

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
```

```python
model = NearestNeighbors(algorithm='brute', metric='cosine')
model.fit(scaled_data)
```

```python
transformer = FunctionTransformer(model.kneighbors)
pipeline = Pipeline(
    [
        ('std_scaler', scaler),
        ('NN', transformer)
    ]
)
params = {'n_neighbors': 5, 'return_distance': False}
pipeline.set_params(NN__kw_args=params)
```
